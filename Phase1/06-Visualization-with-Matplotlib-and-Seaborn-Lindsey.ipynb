### What are the most common last names?

Wait, the last names are part of the full name column... first we'll need to grab the last names!

A few ways we can do this:

```python
x = "here we are, here we go"
```

```python
x.split()
```

```python
# Need to grab last names ... but how?
df['Name'].map(lambda x: x.split(sep=',')[0])
```

```python
ln_list = []
for row in df.index:
    ln = df['Name'].str.split(",")[row][0]
    
    ln_list.append(ln)
```

```python
ln_list[:5]
```

```python
df['Name'].str.split(",")
```

```python
last_list = []
for broken_out_name in df['Name'].str.split(","):
    last_list.append(broken_out_name[0])
```

```python
[broken_out_name[0] for broken_out_name in df['Name'].str.split(",")]
```

```python
df['Last Name'] = pd.Series([name[0] for name in df['Name'].str.split(",")])
```

```python
df.head()
```

`groupby` or `value_counts`?

```python
most_com_lns = df.groupby(by = 'Last Name').count()[['PassengerId']].rename(columns={"PassengerId":"Count"}).sort_values(by="Count", ascending=False).head(8)
```

```python
df['Last Name'].value_counts()
```

Now let's visualize:

```python
most_com_lns
```

```python
# More imports!
import matplotlib.pyplot as plt
```

```python
# Plot a bar graph of the most common last names

most_com_lns.plot(kind='bar', figsize=(10,8))
plt.show()
```

```python
most_com_lns['Count']
```

```python
most_com_lns.index
```

```python
plt.figure(figsize=(10,8))
plt.bar(x=most_com_lns.index, height=most_com_lns['Count'], label='Count')
plt.legend()
plt.show()
```

### Who Paid the Most Fare?

As always - a few ways we can do this: `.loc` versus `sort_values`

```python
most_fare = df.sort_values(by='Fare',ascending=False).head(10)
```

```python
df['Fare'].sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
most_fare
```

```python
df['Fare'].mean()
```

```python
plt.bar(x=most_fare['Name'], height=most_fare['Fare'])
plt.hlines(y=df['Fare'].mean(), xmin=-1, xmax=10, label = "Average Fare Paid", color ='y')
plt.hlines(y=df['Fare'].median(), xmin=-1, xmax=10, label = "Median Fare Paid", color='r')
plt.legend()
plt.xticks(rotation='vertical')
plt.xlabel('Name')
plt.ylabel('Fare (in dollars)')
plt.title('Top Fare Payers on Titanic')
plt.show()
```
